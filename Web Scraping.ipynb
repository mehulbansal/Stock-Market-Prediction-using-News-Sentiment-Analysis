```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.moneycontrol.com/news/tags/nifty.html'
response = requests.get(url)
```

```python
print("response.ok : {} , response.status_code : {}".format(response.ok , response.status_code))
```

```python
from selenium import webdriver
driver = webdriver.Edge()
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
div_tags = soup.find_all("li", {'id': "newslist-"+(str(0))})
div_tags[0].find_all("a")[1].text
```

```python
div_tags = []
for i in range(0,25):
    div_tags.append(soup.find_all("li", {'id': "newslist-"+(str(i))}))
print(div_tags)
```

```python
scraped_headlines = []
for x in div_tags:
    scraped_headlines.append(x[0].find_all("a")[1].text)
print(scraped_headlines)
```
